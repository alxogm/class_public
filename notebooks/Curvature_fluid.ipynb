```python
from classy import Class
import matplotlib.pyplot as plt
from math import pi
```

```python
om_cfld=[-0.1,-0.01,0,0.01,0.1]
```

```python
plt.figure(figsize=(12,12))


for om in om_cfld:
# create instance of the class "Class"
    LambdaCDM_cfld = Class()
    # pass input parameters
    LambdaCDM_cfld.set({'omega_cfld':om,'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842})
    LambdaCDM_cfld.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
# run class
    LambdaCDM_cfld.compute()
    cls = LambdaCDM_cfld.lensed_cl(2500)
    ll = cls['ell'][2:]
    clTT = cls['tt'][2:]
    plt.plot(ll,clTT*ll*(ll+1)/2./pi,label=str(om))
    
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.legend()
plt.savefig("cl.pdf")
```
